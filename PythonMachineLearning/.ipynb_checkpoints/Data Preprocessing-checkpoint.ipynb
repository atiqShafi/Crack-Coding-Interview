建立非数值的或者不可排序的数据到新数据的映射关系

```python
size_mapping = {
    'XL': 3,
    'L': 2,
    'M': 1}

df['size'] = df['size'].map(size_mapping)
df
```

```python
import numpy as np
class_mapping = {label: idx for idx,label in enumerate(np.unique(df['classlabel']))}
df['classlabel'] = df['classlabel'].map(class_mapping)

inverse_class_mapping = {v:k for k,v in class_mapping.items()}
```

## 划分数据集
加载wine数据集
url = https://archive.ics.uci.edu/ml/machine-learning-databases/wine/

```python
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler

df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
df_wine.columns = ['Class Label','Alcohol','Malic acid','Ash','Alcalinity of ash',
                   'Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols',
                   'Proanthocyanins','Color intensity','Hue',"OD280/OD315 of diluted wines","Proline"]
#df_wine.loc[0:20,[col for col in df_wine.columns]]
#type(df_wine)
#df_wine.head()
#print('Class labels',np.unique(df_wine['Class Label']))
#type(df_wine.count)
#help(pd.read_csv)

#from sklearn.cross_validation import train_test_split
X,y = df_wine.iloc[:,1:].values, df_wine.iloc[:,0].values
#type(X)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)

ss = StandardScaler()
ss.fit(X_train)
X_train_std = ss.transform(X_train)
X_test_std = ss.transform(X_test)
```

## 说明归一化和标准化的区别以及应用场景

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler

nr = Normalizer()
ss = StandardScaler()

X1,X2 = df_wine.iloc[:,1],df_wine.iloc[:,2]
X = np.array([X1,X2]).T

XN = nr.fit_transform(X)
XS = ss.fit_transform(X)
print(XN.shape)

plt.scatter(X1,X2,c='r',label='origin')
plt.scatter(XN[:,0],XN[:,1],c='g',label='normalization')
plt.scatter(XS[:,0],XS[:,1],c='b',label='standard')
plt.xlim(-5,20)
plt.ylim(-2,7)
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty='l1',C=0.1)
lr.fit(X_train_std,y_train)
print('Train Accuracy is： ',lr.score(X_train_std,y_train))
print('Test Accuracy is: ',lr.score(X_test_std,y_test))
#lr.coef_
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = plt.subplot(111)
colors = ['blue','green','red','cyan','magenta','yellow','black',
         'pink','lightgreen','lightblue','gray','indigo','orange']

weights,params = [],[]
for c in np.arange(-4,6):
    lr = LogisticRegression(penalty='l1',C=10**c,random_state=0)
    lr.fit(X_train_std,y_train)
    weights.append(lr.coef_[1])
    params.append(10**c)
    
weights = np.array(weights)

for i,c in zip(range(weights.shape[1]),colors):
    plt.plot(params,weights[:,i],
            label=df_wine.columns[i+1],
            color = c)
    
    
plt.axhline(0,color='black',linestyle='--',linewidth=3)
plt.xlim([10**(-5),10**5])
plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.xscale('log')
plt.legend(loc='upper left')
ax.legend(loc='upper center',
         bbox_to_anchor=(1.38,1.03),
         ncol=1,fancybox=True)

plt.show()
```

## Feature Selection SBS Algorithm

```python
from sklearn.base import clone
from itertools import combinations
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score

class SBS():
    def __init__(self,estimator,k_features,
                scoring=accuracy_score,
                test_size=0.25, random_state=1):
        self.scoring = scoring
        self.estimator = clone(estimator)
        self.k_features = k_features
        self.test_size = test_size
        self.random_state = random_state
        
    def fit(self,X,y):
        X_train,X_test,y_train,y_test = train_test_split(X,y,
                                        test_size=self.test_size,
                                        random_state = self.random_state)
        dim = X_train.shape[1]
        self.indices_ = tuple(range(dim))
        self.subsets_ = [self.indices_]
        score = self._calc_score(X_train,y_train,X_test,y_test,self.indices_)
        self.scores_ = [score]
        
        while dim > self.k_features:
            scores = []
            subsets = []
            
            for p in combinations(self.indices_,r=dim-1):
                score = self._calc_score(X_train,y_train,X_test,y_test,p)
                scores.append(score)
                subsets.append(p)
                
            best = np.argmax(scores)
            self.indices_ = subsets[best]
            self.subsets_.append(self.indices_)
            dim -= 1
            
            self.scores_.append(scores[best])
        
        self.k_score_ = self.scores_[-1]
        
        return self
    
    def transform(self,X):
        return X[:,self.indices_]
    
    def _calc_score(self,X_train,y_train,X_test,y_test,indices):
        self.estimator.fit(X_train[:,indices],y_train)
        y_pred = self.estimator.predict(X_test[:,indices])
        score = self.scoring(y_test,y_pred)
        return score
    

        
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
knn = KNeighborsClassifier(n_neighbors=2)
sbs = SBS(knn,k_features=1)
sbs.fit(X_train_std,y_train)

#k_feat = [len(k) for k in sbs.subsets_]
#plt.plot(k_feat, sbs.scores_, marker='o')

plt.plot(np.array(np.arange(X_train_std.shape[1],0,-1)),np.array(sbs.scores_),marker='o')
plt.grid()
plt.xlabel('K features')
plt.ylabel('scores')
plt.ylim([0.7,1.1])
plt.show()


#np.array(np.arange(X_train_std.shape[1],0,-1))
```
